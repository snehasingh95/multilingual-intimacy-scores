```python
import pandas as pd
import numpy as np

import torch
from torch import nn
import numpy as np
from transformers import AutoModel, AutoTokenizer

from torch.optim import Adam
from tqdm import tqdm
```

```python
datapath = '/content/train_copy.csv'
df = pd.read_csv(datapath)
df.head()
```

```python
df = df.loc[df['language'] == "Chinese"]
df.head()
```

```python
df = df.drop('label', axis=1)
df = df.drop('language', axis=1)
df.head()
```

```python
np.random.seed(112)
df_train, df_validate, df_test = \
              np.split(df.sample(frac=1, random_state=42), 
                       [int(.8*len(df)), int(.9*len(df))])
```

```python
print(df_train.shape)
print(df_validate.shape)
print(df_test.shape)
```

```python
df_train.to_csv(r'/content/train.txt', header=None, index=None, sep='\t', mode='a')
df_validate.to_csv(r'/content/validate.txt', header=None, index=None, sep='\t', mode='a')
df_test.to_csv(r'/content/test.txt', header=None, index=None, sep='\t', mode='a')
```
